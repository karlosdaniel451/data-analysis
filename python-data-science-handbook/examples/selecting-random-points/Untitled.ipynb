```python
import numpy as np

mean = [0, 0]
cov = [[1, 2],
       [2, 5]]

X = np.random.multivariate_normal(mean, cov, 100)
X.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set_theme()  # or plot styling

plt.scatter(X[:, 0], X[:, 1]);
```

```python
# Select 20 unique integers in the interval [0, X.shape[0] - 1]
# to select rows in X via fancy indexing.

indices = np.random.choice(X.shape[0], 20, replace=False)
indices
```

```python
selection = X[indices]
print(selection.shape, '\n', selection)
```

```python
plt.scatter(X[:, 0], X[:, 1], alpha=0.3)
plt.scatter(selection[:, 0], selection[:, 1], facecolor='none', s=200);
```

```python
plt.scatter(X[:, 0], X[:, 1], alpha=0.3)
plt.scatter(selection[:, 0], selection[:, 1], facecolor='none', s=2000);
```
